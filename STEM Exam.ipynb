CONFOUNDING VARIABLES 

As mentioned, the crowd on which the volume is measured is not only an independent variable, but would also in this case count as a confounding one; are the people in the crowd the target audience for the specific play or not? Which culture do they come from? (My experience is that watching performances both in the cinema and in the theatre is a very different experience depending on where in the world - the western cinemas and theatres have a tendency to host more quite guests). 

Another confounding variable is whether the loud and cheerful friend group happens to be places close to the mic or not? 

EXPERIMENTAL ERRORS 

Assuming that this is a traditional theater, where the audience doesn't interact with the performing artists, you would also be able to assume that they would not be visible to anyone. Therefore, the way they can show their feelings about the show is by making sounds; and this could be both positive cheers aswell as negative booing. A clear experimental error is therefore only measuring the volume of the sound and not taking into consideration the qualitative outcome of the recording, a negative outburst might give the same volume as a positive one.

Another error would be recording it with only one mic - the further away from the sound a person is sitting, the lower the volume recorded would be.
